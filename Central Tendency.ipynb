```python
descriptive = pd.DataFrame(index=['Mean','Median','Mode','Q1:25%','Q2:50%','Q3:75%','99%','Q4:100%'],columns=quan)
for columns in quan:
    descriptive[columns]["Mean"] = dataset[columns].mean()
    descriptive[columns]["Median"] = dataset[columns].median()
    descriptive[columns]["Mode"] = dataset[columns].mode()[0]
    descriptive[columns]['Q1:25%']=dataset.describe()[columns]['25%']
    descriptive[columns]['Q2:50%']=dataset.describe()[columns]['50%']
    descriptive[columns]['Q3:75%']=dataset.describe()[columns]['75%']
    descriptive[columns]['99%']= np.percentile(dataset[columns],99)
    descriptive[columns]['Q4:100%']=dataset.describe()[columns]['max']
```
